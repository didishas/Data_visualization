```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
housing_df = pd.read_csv('../csv_files/cleaned_properties.csv')
housing_df.sort
```

```python
area_price_df = housing_df.sort_values(by=['area', "price"], ascending=False)
area_price_df['price'] = area_price_df['price'] / 1000
area_price_df['area'] = area_price_df['area'] / 1000000
```

```python
# plt.figure(figsize=(8, 6))


           
sns.scatterplot(x='area', y='price', data=area_price_df)
 
# Tweak using Matplotlib
plt.ylim(0, 200)
plt.xlim(0, None)
```

```python
sns.violinplot(x='area', y='price', data=area_price_df)
```
